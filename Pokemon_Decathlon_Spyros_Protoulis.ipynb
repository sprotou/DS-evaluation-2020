(10 points) Professor Oak suspects that Pokémon in the grass class have a stronger regular attack than those in the rock class. Check if he is right and convince him of your conclusion with statistical arguments. Note: If there are Pokémon that are in both classes, discard them. Note: You may assume that the distribution of the regular attack feature is normal for all Pokémon classes.

```python
pkGrassDf = pkDf[(pkDf['Class 1']=='Grass') | (pkDf['Class 2']=='Grass')]
pkRockDf = pkDf[(pkDf['Class 1']=='Rock') | (pkDf['Class 2']=='Rock')]
```

```python
pkDiscard = pkDf[
   ( (pkDf['Class 1']=='Rock') & (pkDf['Class 2']=='Grass') |
   (pkDf['Class 1']=='Grass') & (pkDf['Class 2']=='Rock'))
]
pkDiscard #we will use this to filter out those that are both classes
```

```python
pkGrassDf2 = pkGrassDf[pkGrassDf['pid'].isin(pkDiscard['pid']) == False] #filter out those that are both Grass and rock
pkRockDf2 = pkRockDf[pkRockDf['pid'].isin(pkDiscard['pid']) == False] #filter out those that are both Grass and rock
```

```python
print('Rock mean atk: ',pkRockDf2['Attack'].mean()
      ,'Rock median atk: ', pkRockDf2['Attack'].median()
      ,' std dev: ' , pkRockDf2['Attack'].std())
print('Grass mean atk: ',pkGrassDf2['Attack'].mean()
      ,'Grass median atk: ', pkGrassDf2['Attack'].median()
      ,' std dev: ' , pkGrassDf2['Attack'].std())
```

In our case, since our dataset contains all the pokemon in existance, comparing the means and medians is enough to prove grass pokemon have WEAKER regular attacks than rock pokemon, since we know that the distributions are normal.

However, if our dataset did not contain all the pokemon but just a sample of the general pokemon population, we would need to do more testing than just comparing the means to show with significant confidence that one class has higher attack than the other.

See below:

```python
stats.ttest_ind(pkGrassDf['Attack'], pkRockDf['Attack'])
```

The p-value indicated the likelihood of the result we received being by chance. 
Given how low it is, it is rather safe to assume that indeed, Grass Pokemon have weaker regular attacks than Rock Pokemon
